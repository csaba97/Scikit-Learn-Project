```python
dat = s1[['Media de\nabsolvire','Nota la\nmatematica','\nNota la\nlimba romana','Unitatea de învățământ', 'Limba și literatura română\r\nScris']]
```

```python
le = LabelEncoder()
dat['Unitatea de învățământ'] = le.fit_transform(dat['Unitatea de învățământ'])
```

```python
dat['Unitatea de învățământ']
```

```python
X = dat.loc[:, dat.columns != 'Limba și literatura română\r\nScris']
```

```python
X.shape
```

```python
y = s1['Limba și literatura română\r\nScris'].values.reshape(-1,1)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train) #training the algorithm
```

```python
#To retrieve the intercept:
print(regressor.intercept_)

#For retrieving the slope:
print(regressor.coef_)
```

```python
y_pred = regressor.predict(X_test)
```

```python
res = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
```

```python
res
```

```python
print('Mean Squared Error:', mean_squared_error(y_test, y_pred)) 
```

```python
accuracy_score('Limba și literatura română\r\nScris', y_pred)
```
