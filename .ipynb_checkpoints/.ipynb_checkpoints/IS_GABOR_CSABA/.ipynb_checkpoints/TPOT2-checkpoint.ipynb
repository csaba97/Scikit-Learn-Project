```python
# in this notebook, we'd like to predict the label 'romana final'
# this is a regression problem, because our label is a continuos value
df = pd.read_csv('note.csv')

df = df.reset_index()

X = df.drop(['index','judet8','scoala','media la admitere','media teze nationale', 'nume', 'materna oral','materna nota',
             'materna contestatie','materna final',
            'disciplina profil nota', 'disciplina profil contestatie', 'optional nota', 'optional contestatie',
            'media','rezultatul final','romana contestatie','romana final','romana nota','pozitia in ierarhie pe tara',
            'pozitia in ierarhie pe judet','limba materna8 nota', 'unitatea de invatamant','judetul','limba materna'],
            
            axis=1)
Y = df[['romana final']]
```

```python
X['limba moderna nota'].fillna(4.0, inplace=True)#grade 4.0(highest grade) for those who have a language exam
```

```python
for i,c in enumerate(X.columns):
        if X[c].dtype == 'object':
            lbl = LabelEncoder()
            lbl.fit(list(X[c].values))
            X[c] = lbl.transform(list(X[c].values)) 
X.head()
```

```python
for i,c in enumerate(Y.columns):
        if Y[c].dtype == 'object':
            lbl = LabelEncoder()
            lbl.fit(list(Y[c].values))
            Y[c] = lbl.transform(list(Y[c].values)) 
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20, random_state = 40) #same seed to compare results
```

```python
tpot = TPOTRegressor(verbosity=2, max_time_mins=5)
tpot.fit(X_train, Y_train)
print(tpot.score(X_test, Y_test))
```
