```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv('note.csv')
df = df.reset_index()
```

```python
X = df.drop(['index','romana contestatie','limba materna','materna oral','materna nota','materna contestatie','materna final',
            'limba moderna nota','disciplina profil nota', 'disciplina profil contestatie', 'optional nota', 'optional contestatie',
            'media','rezultatul final','romana final','romana nota','pozitia in ierarhie pe tara', 'pozitia in ierarhie pe judet'],
            axis=1)
Y = df['romana final'].values.reshape(-1,1)      
```

```python
X.isna().sum()
```

```python
X.keys()
```

```python
categorical_features = []
categorical_features_indices = []
for i,c in enumerate(X.columns):
        if X[c].dtype == 'object':
            categorical_features.append(c)
            categorical_features_indices.append(i)
            lbl = LabelEncoder()
            lbl.fit(list(X[c].values))
            X[c] = lbl.transform(list(X[c].values))
```

```python
print X.shape
print Y.shape
```

```python
categorical_features
categorical_features_indices
```

```python
cat = OneHotEncoder()
X = cat.fit_transform(X).toarray()
```

```python
X
```

```python
#deprecated
onehotencoder = OneHotEncoder(categorical_features = categorical_features_indices)
X = onehotencoder.fit_transform(X).toarray()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.20) 
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))
X = scaler.fit_transform(X)

regressor = LinearRegression()  
regressor = LinearRegression()  
regressor.fit(X_train, Y_train) #training the algorithm
```

```python
Y_pred = regressor.predict(X_test)
```

```python
mean_squared_error(Y_test, Y_pred)
```

```python
mean_absolute_error(Y_test, Y_pred)
```

```python
regressor.score(X_test, Y_test)
```

```python
#feautre importance
coefs = regressor.coef_[0]
indices = np.argsort(coefs)[::-1]

plt.figure()
plt.title("Feature importances (Logistic Regression)")
plt.bar(range(10), coefs[indices[:10]],
       color="r", align="center")
plt.xticks(range(10), X, rotation=45, ha='right')
plt.subplots_adjust(bottom=0.3)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Imputer
my_pipeline = make_pipeline(Imputer(), RandomForestRegressor())
from sklearn.model_selection import cross_val_score
scores = cross_val_score(my_pipeline, X, Y, scoring='neg_mean_squared_error')
print(scores)
```

```python
model = RandomForestRegressor(n_jobs=-1)
# Try different numbers of n_estimators - this will take a minute or so
estimators = np.arange(10, 100, 10)
scores = []
for n in estimators:
    model.set_params(n_estimators=n)
    model.fit(X_train, Y_train.ravel())
    scores.append(model.score(X_test, Y_test))
plt.title("Effect of n_estimators")
plt.xlabel("n_estimator")
plt.ylabel("score")
plt.plot(estimators, scores)
```

```python
#random forest
rfc = RandomForestRegressor(n_estimators=10)
rfc.fit(X_train, Y_train.ravel())

feature_importances = pd.DataFrame(rfc.feature_importances_,
                                   index = X_train.columns,
                                    columns=['importance']).sort_values('importance',
                                                                        ascending=False)

feature_importances
```
