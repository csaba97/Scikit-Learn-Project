```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#this notebook is only for plots, does not contain experiments
df = pd.read_csv('note.csv')
df = df.reset_index()
```

```python
df.head()
```

```python
def show_plot_for(col_name, width=5, height=5, logy=False):
    plt.clf()
    df.groupby(col_name).size().plot(kind='bar', figsize=(width,height), logy=logy)
    plt.show()
```

```python
show_plot_for('judet8', 10, 10)
```

```python
show_plot_for('optional8')
```
